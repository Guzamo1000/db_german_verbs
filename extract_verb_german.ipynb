```python
def get_text(section):
    tr=section.find_all("tr")
    td=[]
    for t in tr:
        dir_={}
        td_=t.find_all("td")
        
        # print(f"len td: {len(td)}")
        if len(td_)>1:
            # dir_['conjugated_verbs']="None"
            dir_["type"]=td_[0].text
            dir_['conjugated_verbs']=td_[1].text
            td.append(dir_)
        else: 
            dir_['conjuge']=td_[0].text
            td.append(dir_)
    return td
```

```python
section=soup.find_all("section", class_="rBox rBoxWht")
print(len(section))
if len(section)==0: 
    print("no identification")
total_data=[]
se=len(section)
if len(section)>6: se=6

for s in range(1,se+1):
    # print(s)
    div_data=[]
    data_section={}
    div=section[s].find_all("div", class_="vTbl")
    if s==1:
        data_section['section']="simple"
        data_div=[]
        for d in div:
            div_dic={}
            div_dic['tense']=d.find("h2").text
            div_dic['content']=get_text(d)
            # print(f"div data: {div_data}")
            data_div.append(div_dic)
            # data_section['content_section']=div_data
    else:
        data_section['section']=section[s].find("h2").text
        data_div=[]
        for d in div:
            div_dic={}
            div_dic['tense']=d.find("h3").text
            div_dic['content']=get_text(d)
            # print(f"div data: {div_data}")
            data_div.append(div_dic)
    data_section['content_section']=div_data
    # print(f" data section: {data_section}")
    total_data.append(data_section)
print(total_data)
```
